```python
raw_data = pd.read_csv(datafile, skiprows=8)
raw_data
```

Remove unnecessary rows.

```python
df = raw_data.iloc[8:]
```

```python
df
```

```python
df.set_index(df.columns[0],inplace=True)
df
```

```python
print(df.columns)

def clean_column_name(col_name):
    col = col_name.split(" ")
    col.pop()
    return "_".join(col)

update_columns = []
suffix = ["Total","Men","Women"]
for i,each in enumerate(df.columns):
    if "Unnamed" not in each:
        current_column = clean_column_name(each)

    update_columns.append(f"{current_column}_{suffix[i%len(suffix)]}")


print(update_columns)
```

```python
df.columns = update_columns
df
```

Create Copy of Dataframe

```python
df_2 = df.copy()
```

```python
def get_hierarchy_number(value):
    noc = value.split(" ")[0]
    return len(noc)
df_2["Hierarchy_Num"] = df_2.index.to_series().apply(get_hierarchy_number)
```

```python
df_2
```

```python
def split_dataframes(master_df,spliton="Total"):
    cols = [col for col in master_df.columns if spliton in col]
    cols.append("Hierarchy_Num")
    return master_df[cols]

total_df = split_dataframes(df_2,"Total")
```
